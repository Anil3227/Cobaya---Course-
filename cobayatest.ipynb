```python
from getdist import MCSamples, plots
import numpy as np

# Load your chain file
data = np.loadtxt('/Users/anilkandel/cobaya/chains/test.1.txt')

# Uniform weights
weights = np.ones(data.shape[0])

# Extract required parameters (correct column indexes)
H0       = data[:, 4]
Omega_m  = data[:, 11]
w0       = data[:, 7]
wa       = data[:, 8]

# Stack them into samples array
params = np.column_stack([H0, Omega_m, w0, wa])

# Names used by GetDist internally
param_names  = ['H0', 'Omega_m', 'w0', 'wa']
# LaTeX labels for axes
param_labels = [r'H_0', r'\Omega_m', r'w_0', r'w_a']

# Create GetDist sample object
samples = MCSamples(samples=params,
                    weights=weights,
                    names=param_names,
                    labels=param_labels)

# Plot settings
g = plots.getSubplotPlotter(width_inch=5.0)
g.settings.figure_legend_frame = True
g.settings.alpha_filled_add = 0.6
g.settings.axes_labelsize = 12
g.settings.legend_fontsize = 10

# Triangle plot for all four parameters
g.triangle_plot([samples],
                ['H0', 'Omega_m', 'w0', 'wa'],
                filled=True,
                contour_colors=['darkblue'],
                legend_loc='upper right',
                title_limit=1)

# Save the figure
g.export("H0_Omega_m_w0_wa.pdf")
```
